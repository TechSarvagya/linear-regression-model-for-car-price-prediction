```python
def dummies(x,dtf):
 temp=pd.get_dummies(dtf[x], drop_first=True)
 dtf.drop(x,axis=1,inplace=True)
 dtf=pd.concat([dtf,temp],axis=1)
 return dtf
car_data['owner']=car_data['owner'].replace({'First Owner':1, 'Second Owner':2,'Third Owner':3,'Fourth & Above Owner':4})
car_data=dummies('fuel',car_data)
car_data=dummies('seller_type',car_data)
car_data.drop(car_data[(car_data['owner']=='Test Drive Car')].index,axis=0, inplace=True)
    
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def build_model(X,Y):
 X= sm.add_constant(X)
 lm=sm.OLS(Y,X).fit()
 print(lm.summary())
 return lm

def check_vif(X):
 vif=pd.DataFrame()
 vif['Features']=X.columns
 vif['VIF']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
 vif=vif.sort_values(by='VIF',ascending=False)
 return vif
```
